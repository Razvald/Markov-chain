Для начала, необходимо получить текст и токенизировать его.

```python
# Import block

import os
import re
import random
import pandas as pd
```

```python
# Tokenizer

# Файл содержит 3 столбца: writer, poem, text
# Нужно оставить только 3 столбец
folder = "./data/samples/"
corpus = ""

for filename in os.listdir(folder):
    if filename.endswith(".txt"):
        with open(os.path.join(folder, filename), "r", encoding="utf-8") as f:
            corpus += " " + f.read()

# Имея список текстов, нужно объединить их и составить токенизацию
def clean_text(text: str) -> str:
    text = text.lower()
    text = re.sub(r"[^а-яa-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

corpus = clean_text(corpus)
tokens = corpus.split()
```

```python
# Model

# Необходимо превратить одну строку в структуру ключа и его списка, под каждое слово в строке.
transition = {}

for i in range(len(tokens) - 2):
    words = tokens[i], tokens[i + 1]
    next_word = tokens[i + 2]

    if words not in transition:
        transition[words] = []

    transition[words].append(next_word)
```

```python
def gen_loop(transition: dict, count: int) -> str:
    w1, w2 = random.choice(list(transition.keys()))

    result = [w1, w2]

    for _ in range(count - 2):
        next_words = transition.get((w1, w2))
        if not next_words:
            break

        w3 = random.choice(next_words)
        result.append(w3)

        w1, w2 = w2, w3

    return " ".join(result)

text = gen_loop(transition, 50)
print(text)
```
