Весь проект укладывается в несколько этапов:
- Токенизация текста
- Реализация модели
- Создание интерфейса

Для начала, необходимо получить текст и токенизировать его.

```python
# Import block

import os
import re
import random
```

```python
# Tokenizer

folder = "./data/samples/"
corpus = ""

for filename in os.listdir(folder):
    if filename.endswith(".txt"):
        with open(os.path.join(folder, filename), "r", encoding="utf-8") as f:
            corpus += " " + f.read()

def tokenize_with_starts(text: str):
    starts = set()

    raw_tokens = re.findall(r"[А-Яа-яA-Za-z]+|[.!?]", text)

    tokens = []
    new_sentence = True

    for tok in raw_tokens:
        if tok in ".!?":
            new_sentence = True
            continue

        word = tok.lower()
        tokens.append(word)

        if new_sentence:
            starts.add(word)
            new_sentence = False

    return tokens, starts

tokens, start_words = tokenize_with_starts(corpus)
```

```python
# Model

transition = {}

for i in range(len(tokens) - 2):
    key = (tokens[i], tokens[i+1])
    next_word = tokens[i+2]

    if key not in transition:
        transition[key] = []

    transition[key].append(next_word)

start_pairs = []

for i in range(len(tokens) - 1):
    if tokens[i] in start_words:
        start_pairs.append((tokens[i], tokens[i+1]))
```

```python
# Generator

def gen_loop(transition: dict, start_pairs: list, count: int) -> str:
    w1, w2 = random.choice(start_pairs)
    result = [w1.capitalize(), w2]

    for _ in range(count - 2):
        next_words = transition.get((w1, w2))
        if not next_words:
            break

        w3 = random.choice(next_words)
        result.append(w3)

        w1, w2 = w2, w3

    return " ".join(result) + '.'
```

```python
# Run

print(gen_loop(transition, start_pairs, 30))
```
