Для начала, необходимо получить текст и токенизировать его.

```python
# Import block

import re
import random
import pandas as pd
```

```python
# Tokenizer

# Файл содержит 3 столбца: writer, poem, text
# Нужно оставить только 3 столбец
data = pd.read_csv('./data/samples/poems.csv', usecols=['text'])
data.dropna(inplace=True)

# Имея список текстов, нужно объединить их и составить токенизацию
corpus = " ".join(data['text'])

def clean_text(text: str) -> str:
    text = text.lower()
    text = re.sub(r"[^а-яa-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

corpus = clean_text(corpus)
tokens = corpus.split()
```

```python
# Model

# Необходимо превратить одну строку в структуру ключа и его списка, под каждое слово в строке.
transition = {}

for i in range(len(tokens) - 1):
    word = tokens[i]
    next_word = tokens[i + 1]

    if word not in transition:
        transition[word] = []

    transition[word].append(next_word)
```

```python
# Теперь когда у нас на руках есть словарь с вершинами и ребрами, пора составить генерацию
def gen(index: str, dict: dict[str, list[str]], count: int):
    if count == 0:
        return

    words = dict.get(index, [])
    if not words:
        return

    next_word = random.choice(words)
    print(index, end=' ')

    gen(next_word, dict, count - 1)

index = random.choice(list(transition.keys()))
count = 10
gen(index, transition, count)
```
