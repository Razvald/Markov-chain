Для начала, необходимо получить текст и токенизировать его.

```python
# Import block

import os
import re
import random
import pandas as pd
```

```python
# Tokenizer

# Файл содержит 3 столбца: writer, poem, text
# Нужно оставить только 3 столбец
folder = "./data/samples/"
corpus = ""

for filename in os.listdir(folder):
    if filename.endswith(".txt"):
        with open(os.path.join(folder, filename), "r", encoding="utf-8") as f:
            corpus += " " + f.read()

# Имея список текстов, нужно объединить их и составить токенизацию
def clean_text(text: str) -> str:
    text = text.lower()
    text = re.sub(r"[^а-яa-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

corpus = clean_text(corpus)
tokens = corpus.split()
```

```python
# Model

# Необходимо превратить одну строку в структуру ключа и его списка, под каждое слово в строке.
transition = {}

for i in range(len(tokens) - 2):
    words = tokens[i], tokens[i + 1]
    next_word = tokens[i + 2]

    if words not in transition:
        transition[words] = []

    transition[words].append(next_word)
```

```python
# Теперь когда у нас на руках есть словарь с вершинами и ребрами, пора составить генерацию
def gen(index: tuple, dict: dict, count: int):
    if count == 0:
        return

    words = dict.get(index, [])
    if not words:
        return

    next_word = index[1], random.choice(words)
    print(index[0], end=' ')

    gen(next_word, dict, count - 1)

index = random.choice(list(transition.keys()))
count = 50
gen(index, transition, count)
```
